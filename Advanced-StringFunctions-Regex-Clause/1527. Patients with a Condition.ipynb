<p>Write a solution to find the patient_id, patient_name, and conditions of the patients who have Type I Diabetes. Type I Diabetes always starts with <code>DIAB1</code> prefix.</p>

<p>Return the result table in <strong>any order</strong>.</p>

Patients table:

| patient_id | patient_name | conditions   |
+------------+--------------+--------------+
| 1          | Daniel       | YFEV COUGH   |
| 2          | Alice        |              |
| 3          | Bob          | DIAB100 MYOP |
| 4          | George       | ACNE DIAB100 |
| 5          | Alain        | DIAB201      |
+------------+--------------+--------------+
<strong>Output:</strong> 
+------------+--------------+--------------+
| patient_id | patient_name | conditions   |
+------------+--------------+--------------+
| 3          | Bob          | DIAB100 MYOP |
| 4          | George       | ACNE DIAB100 | 
+------------+--------------+--------------+
<strong>Explanation:</strong> Bob and George both have a condition that starts with DIAB1.
</pre>

```python
# Import PySpark pandas API and data types for schema definition
import pyspark.pandas as ps
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

# Sample data for Patients table
patients_data = [
    (1, "Daniel", "YFEV COUGH"),
    (2, "Alice", ""),
    (3, "Bob", "DIAB100 MYOP"),
    (4, "George", "ACNE DIAB100"),
    (5, "Alain", "DIAB201")
]

# Define schema for Patients table
patients_schema = StructType([
    StructField("patient_id", IntegerType(), False),
    StructField("patient_name", StringType(), False),
    StructField("conditions", StringType(), False)
])

# Create Spark DataFrame from sample data and schema
df_patients = spark.createDataFrame(patients_data, schema=patients_schema)

# Convert Spark DataFrame to pandas-on-Spark DataFrame
df_patients_pandas = df_patients.pandas_api()

# Register DataFrame as a temporary SQL view
df_patients.createOrReplaceTempView("Patients")

# Display the DataFrame in Databricks notebook
display(df_patients)
```

```python
# Display patients who have a condition starting with 'DIAB1' (Type I Diabetes)
display(df_patients.where(df_patients.conditions.contains("DIAB1")))
```

```python
# Filter patients whose 'conditions' contain 'DIAB1' (Type I Diabetes)
df_patients_pandas[df_patients_pandas["conditions"].str.contains("DIAB1")].reset_index(
    drop=True
)
```

```python
%sql
-- Select patients who have a condition starting with 'DIAB1' (Type I Diabetes)
select
  *
from
  Patients
where
  conditions like '%DIAB1%'
```
